```python
from TF_IDF import TF_IDF
import numpy as np


trainFolderPath = r"G:\My Drive\EPR_420\Code\Content Classification\Testing Document Templates\A) Latin Language PDFs\B) Indonesian PDFs\Training"
languageDictionaryPath = r"G:\My Drive\EPR_420\Code\main\dictionaries\indonesian_words_dictionary.json"

# y_label generated within constructor 
tf_idf = TF_IDF(trainFolderPath, languageDictionaryPath)
print(tf_idf.Get_y_labels_train())
tf_idf.CalculateVocabList()
tf_idf.CalculateTermFreqTable()
tf_idf.Calculate_TF_IDF_Train()


```

Multiple PDFs Test:

```python
from KNN import KNN

tf_idf_train = tf_idf.Get_TF_IDF_Train()

testFolderPath = r"G:\My Drive\EPR_420\Code\Content Classification\Testing Document Templates\A) Latin Language PDFs\B) Indonesian PDFs\Testing"
tf_idf_testPDFs, Actual_y_Labels = tf_idf.Get_Test_TF_IDF_Multiple_PDFs(testFolderPath)

print(tf_idf_testPDFs.shape)


knn = KNN(3, 8)
knn.load(tf_idf_train, tf_idf.Get_y_labels_train())


for i in range(len(tf_idf_testPDFs)):
    print("Doc#:", i, "Actual_y_Labels", Actual_y_Labels[i])
    knn._predict(tf_idf_testPDFs[i])
    print("")
    
    # print("Actual_y_Labels", Actual_y_Labels[i])
```
